```python
import cv2
import matplotlib.pyplot as plt
```

```python
imgA = cv2.imread("img/foto1A.jpg")
imgA = imgA[..., ::-1]
imgB = cv2.imread("img/foto1B.jpg")
imgB = imgB[..., ::-1]
```

```python
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
plt.title('Foto A'), plt.xticks([]), plt.yticks([])
imgplot = plt.imshow(imgA)

a = fig.add_subplot(1, 2, 2)
plt.title('Foto B'), plt.xticks([]), plt.yticks([])
imgplot = plt.imshow(imgB)
plt.show()
```
